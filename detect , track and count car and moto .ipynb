```python
import cv2
from tracker import *
import os
#os.chdir('C:\\Users\\Mostafa\\Downloads\\Artificial Intelligence\\Computer Vsion')

tracker = EuclideanDistTracker()
cap = cv2.VideoCapture('C:\\Users\\Mostafa\\Downloads\\Artificial Intelligence\\Computer Vsion\\highway.mp4')

object_detector = cv2.createBackgroundSubtractorMOG2(history=50,varThreshold=100)

while True :
    ret , frame = cap.read()
    heigh,width,_ = frame.shape
    
    roi = frame[340:720,500:800]
    mask = object_detector.apply(roi)
    
    _,mask = cv2.threshold(mask, 254, 255, cv2.THRESH_BINARY)
    contours , _ = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    detections = []
    for contour in contours :
        area = cv2.contourArea(contour)
        if area > 100 :
            x , y , h , w = cv2.boundingRect(contour)
            
            detections.append([x,y,h,w])
    boxes_ids = tracker.update(detections)
    for box_id in boxes_ids :
        x,y,w,h,id = box_id
        cv2.putText(roi, str(id), (x,y), cv2.FONT_HERSHEY_COMPLEX, 2, (0,255,0),2)
        cv2.rectangle(roi, (x,y), (x+w,y+h),(0,255,0),3)
    cv2.imshow('roi',roi)
    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    
    key = cv2.waitKey(30)
    if key == 27 :
        break
cap.release()
cv2.destroyAllWindows()
```
